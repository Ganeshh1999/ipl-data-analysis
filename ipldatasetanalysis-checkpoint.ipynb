```python
ipl_matches = pd.read_csv(r'C:\Users\ganesh\Downloads\ipldata\matches.csv')
ipl_matches
```

# RCB vs CSK wins percentages

```python
df = ipl_matches[((ipl_matches['team1']=="Royal Challengers Bangalore") & (ipl_matches['team2']=="Chennai Super Kings")) | ((ipl_matches['team2']=="Royal Challengers Bangalore") & (ipl_matches['team1']=="Chennai Super Kings"))] 
rcb_wins=df[df['winner']=="Royal Challengers Bangalore"].count()
csk_wins=df[df['winner']=="Chennai Super Kings"].count()
rcb_wins


```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'rcb', 'csk'
sizes = [rcb_wins['id'], csk_wins['id']]
colors = ['gold', 'lightskyblue']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

# rcb wins against csk (first batting vs first fielding)

```python
df1 = df[((df['toss_decision']=="bat")&(df['toss_winner']=="Royal Challengers Bangalore"))|((df['toss_winner']=="Chennai Super Kings")&(df['toss_decision']=="field"))]
rcb_wins_bat = df1[df1['winner']=="Royal Challengers Bangalore"].count()
rcb_wins_bat
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'rcb_wins_field', 'rcb_wins_bat'
sizes = [rcb_wins['id']- rcb_wins_bat['id'], rcb_wins_bat['id']]
colors = ['gold', 'lightskyblue']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

# Rcb choosing after winning the toss

```python
df1 = ipl_matches[(ipl_matches['toss_winner']=="Royal Challengers Bangalore") & (ipl_matches['toss_decision']=="field")]
df2 = ipl_matches[(ipl_matches['toss_winner']=="Royal Challengers Bangalore") & (ipl_matches['toss_decision']=="bat")]
```
